### Tuples:
Declared with () instead of [] in lists

Main difference between list and tuple is that data declared as tuples is immutable i.e. it can't be changed as data in list can be changed.

Individual tupple accessed through the same way as list

```python
t = (1,3,5)
print(t)
type(t)

# Will show error
# t[2] = 55
```

### File pointer

```python
fp = open('calender.py', 'r')
type(fp)
#print(fp.read())
```
